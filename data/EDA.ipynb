```python
df = pd.read_csv(path+'nepsedata.csv')
### find all the data with the company_name as SHLB
def findbycompanyname(df,company_name):
    #convert the data to datetime
    df['date']=pd.to_datetime(df['date'])
    df=df[df['company_name']==company_name]
    plt.figure(figsize=(20,12))    
    plt.plot(df['date'],df['closing_price'])
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.title(company_name)

data=findbycompanyname(df,'SMJC')

```

```python
##total traded qunatity
df = pd.read_csv(path+'nepsedata.csv')
df['date']=pd.to_datetime(df['date'])

def findbycompanyname_totaltradedquantity(df,company_name):
    df=df[df['company_name']==company_name]
    plt.figure(figsize=(20,12))
    plt.subplot(2, 2, 2)
    plt.plot(df['date'],df['total_traded_quantity'])
    plt.xlabel('Date')
    plt.ylabel('Total Traded Quantity(kitta)')
    plt.title(company_name)

data=findbycompanyname_totaltradedquantity(df,'SHLB')
```

```python
### total traded value


def cleanDataFirst():
    df = pd.read_csv(path+'nepsetodayRealtime.csv')
    #delete the row if company_name is "company_name"
    df = df[df['company_name']!="company_name"]
    columns_list =['confidence', 'open_price', 'lowest_price',
       'highest_price', 'closing_price', 'VWAP', 'total_traded_quantity',
       'Previous_closing', 'total_traded_value', 'total_trades', 'difference',
       'range', 'difference_percentage', 'range_percentage', 'VWAP_percentage',
       '52_weeks_high', '52_weeks_low']
    for column in columns_list:
        df[column] = df[column].apply(lambda x: float(x.replace(",","")) if type(x)==str else x)    
    return df   
    
def realtimeTime(company_name):
    df=cleanDataFirst()
    df=df[df['company_name']==company_name]
    plt.figure(figsize=(20,12))
    plt.subplot(2, 2, 2)
    plt.plot(df['time'],df['total_traded_value'])
    plt.xlabel('Time')
    plt.ylabel('Total Traded Value')
    plt.title(company_name)


realtimeTime('SHLB')


    
    
```
